```python
def nb(m, s, l):
    res = (-1.0)**(l - 1.0) / l * (
            torch.exp( m * l + 0.5 * s**2 * (l ** 2) + log_ndtr(-m / s - s * l)) + \
            torch.exp(-m * l + 0.5 * s**2 * (l ** 2) + log_ndtr( m / s - s * l))
        )
    return res
```
